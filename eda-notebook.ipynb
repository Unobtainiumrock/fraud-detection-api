# Random Cut Forest For Anomaly Detection (Unsupervised Learning)

The data we're working with is a high-dimensional time series and realistically speaking, we won't have all data neatly labeled (see more in README). This is where anomaly detection comes into play and it can be achieved with a Random Cut Forest.

## RCF

A Random Cut Forest is an algorithm primarily used for anomaly detection, introduced by Amazon. The math behind a Random Cut Forest can be broken down into a series of steps, with the key idea being to build a collection of trees by recursively applying random hyperplane cuts to the data points, rather than traditional binary tree splits. Each tree represents a set of cuts that partition the data. Anomalies are characterized as those that are easily isolated by fewer cuts, meaning that they fall into their own partition early in the process.

Since these cuts aren't constrained by a specific decision tree structure, they can capture more complex data distributions than traditional binary trees. After constructing the forest of trees, the anomaly score of a point is computed based on how isolated it becomes within the forest, with more isolated points being more likely to be anomalies.

By applying random hyperplane cuts in multiple trees and then averaging the results across the forest, RCF effectively models the distribution of normal points and highlights anomalies that are easily separated by fewer cuts, capturing complex patterns in the data.

## RCF Advantages

Random Cut Forests excel in datasets where distributions change over time because they are inherently adaptive and can efficiently handle data drift or evolving patterns.

### Incremental Leanring

RCFs can be updated dynamically as new data arrives, without needing to completely retrain the model from scratch. This means that they can adjust to new patterns or changes in the distribution of data over time. For example, in a scenario where the characteristics of normal behavior shift gradually, the model can adapt by incorporating new data and "forgetting" outdated patterns.

### No Underlying Assumptions About Data Distribution

Since RCFs do not rely on specific assumptions about the underlying distribution of the data, they are well-suited to handle evolving distributions. Unlike statistical models that assume stationarity (i.e., that the data distribution remains constant over time), RCFs work by isolating data points using random hyperplane cuts, which allows them to capture changes in the distribution naturally, without needing to conform to a fixed model.


### Forest Diversity

The randomness inherent in the cuts, combined with building multiple trees in the forest, allows RCFs to maintain a diverse set of models. This diversity helps the model remain robust to different types of changes in data distribution, as different trees will capture different aspects of the data. When the distribution shifts, some trees may adapt more quickly, while others may still capture prior patterns, providing a balanced detection mechanism across both old and new behaviors.

## Characterizing Anomalies

- **Fewer Cuts for Anomalies:** Points that are anomalies typically require fewer cuts to be isolated from the rest of the data. This is because anomalies tend to be far from other points or in sparse regions of the data space. When we make random cuts, such points often get separated early on since there are fewer neighboring points close to them.

- **More Cuts for Normal Points:** Normal points, which are clustered together with many nearby points, require more cuts to be isolated because they are embedded within dense regions. Random cuts will have to continue partitioning the space to finally separate them from their close neighbors.
